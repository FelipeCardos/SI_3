## 1. Processamento dos dados

(remover) Descrever e comentar os passos na leitura e processamento dos dados 

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text

#leio o CSV de treino e uso o sep="," pois como está na
#pl não funcionava aqui
df = pd.read_csv("train.csv", sep=",")
df
```

```python
#classes que identifiquei que fariam um treino mais preciso
#sao 20(0-19)
col_indexs=[3,12,17]
#nome das colunas 
col_names=np.array(df.columns)[col_indexs]
#valor da classe
x=df.values[:,col_indexs]
#resultado dado aquela classe
y=df.values[:,19]
#converto de float para int
y=y.astype('int')
print(x)
print(".....--------")
print(y)
```

```python
#vejo os atributos para as classes anteriors (12,17)
df_x=df[col_names]
#converto para 1 ou 0
df_xDum = pd.get_dummies(df_x, columns=col_names)
#cria array dos atributos convertidos para 1 ou 0
col_names_dum=np.array(df_xDum.columns)

df_xDum
```

```python
l= df_xDum.values
l
```

```python
sk_dt = DecisionTreeClassifier(criterion="entropy")
sk_dt.fit(l, y)
```

```python
#compara a previsão com o caso real, e com as 2 classes foi
#possivel ter 100% de acerto, por isso usei as classes 12 e 17
preds = sk_dt.predict(l)
for i in range(df_xDum.shape[0]):
    print(f"Classe real: {y[i]}; Previsão: {preds[i]} ")
print("\nNúmero Total de más predições: ", sum(preds!=y))
```

```python
#print da arvore
tree = export_text(sk_dt, feature_names=list(col_names_dum))
print(tree)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_train)
print(y_train)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(X_train,y_train)
```

```python
dtc.score(X_test, y_test)
```
